```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
print('git-r-done')
```

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source, 'lxml')

codes_list=[]
borough_list=[]
neighborhood_list=[]
i=1
for tag in soup.table.find_all('td'):
    if i == 1:
        codes_list.append(tag.text)
    if i == 2:
        borough_list.append(tag.text)
    if i == 3: 
        neighborhood_list.append(tag.text)  
       
    i = i+1
    if i==4:
        i=1

df = pd.DataFrame()
df['PostalCode'] = codes_list
df['Borough'] = borough_list
df['Neighborhood'] = neighborhood_list
df = df.replace('\n',' ', regex=True)
df.drop(df[df['Borough']=="Not assigned"].index,axis=0, inplace=True)
df1=df.groupby("PostalCode").agg(lambda x:', '.join(set(x)))
df1.loc[df1['Neighborhood']=="Not assigned",'Neighborhood']= \
df1.loc[df1['Neighborhood']=="Not assigned",'Borough']

df1

```
